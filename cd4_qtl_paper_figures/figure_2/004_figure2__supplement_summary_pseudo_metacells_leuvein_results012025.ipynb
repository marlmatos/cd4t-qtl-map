```python
import os
import pandas as pd
import anndata
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
resolution=["34"]

for res in resolution:
    # Inputs
    results="/gchm/cd4_CellRegMap/001_preprocessing/results_01312025/prep_phenotype_vector/Data/pseudobulk_metacells/"
    
    # Sample mapping file
    sample_mapping_file = results+'Pseudobulk_per_donor_all_cells_all_conditions_Leiden_res'+res+".h5"
    
    # Load AnnData object from HDF5 file
    adata = anndata.read_h5ad(sample_mapping_file)
    print(adata)
    
    # Create stats dataframe
    stats = pd.DataFrame({
        'Cell_ID': adata.obs.index.tolist(),
        'Number_of_Cells': adata.obs['Number_of_cells'].tolist(),
        'Sample': adata.obs['scRNA_Sample_Name'].astype(str).tolist(),
    })    
    
    # Compute the average number of cells per pseudocell, per sample
    grouped = stats.groupby(['Cell_ID', 'Sample'])['Number_of_Cells'].mean().reset_index()
    
    # Group by 'Sample' to get the mean number of cells per pseudocell for each sample
    average_cells_per_pseudocell = grouped.groupby('Sample')['Number_of_Cells'].mean().reset_index()
    average_cells_per_pseudocell = average_cells_per_pseudocell.rename(columns={'Number_of_Cells': 'Mean_Number_of_Pseudocells'})
    
    # Compute the total number of cells per donor (Sample)
    total_cells_per_donor = stats.groupby('Sample')['Number_of_Cells'].sum().reset_index()
    total_cells_per_donor = total_cells_per_donor.rename(columns={'Number_of_Cells': 'Mean_Number_of_Cells'})
    
    # Merge the DataFrames
    merged_stats = total_cells_per_donor.merge(average_cells_per_pseudocell, on='Sample')
    
    # Create the scatter plot
    plt.figure(figsize=(6, 6))
    plt.scatter(merged_stats['Mean_Number_of_Cells'], 
                merged_stats['Mean_Number_of_Pseudocells'],
                color='#1f77b4', alpha=0.8, s=25)
    
    # Add labels and title
    res1 = int(res)
    plt.ylabel('Average Number of Cells per Pseudocell', fontsize=14)
    plt.xlabel('Total Number of Cells per Donor', fontsize=14)
    plt.title(f'Metacell Pseudobulk for Leiden Res: {res1 / 10:.1f}', fontsize=14)
    
    # Add gridlines for better readability
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Save figure
    plt.show()
```

```python
import anndata
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# List of resolutions you want to compare
resolutions = ["34", "4", "45", "5"]  # example – change to what you actually have

results = "/gchm/cd4_CellRegMap/001_preprocessing/results_01312025/prep_phenotype_vector/Data/pseudobulk_metacells/"

# Set up figure and colormap
plt.figure(figsize=(6, 6))
cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(resolutions)))

# Store summary text per resolution to annotate later
summary_lines = []

for res, color in zip(resolutions, colors):
    # Input file
    sample_mapping_file = results + f"Pseudobulk_per_donor_all_cells_all_conditions_Leiden_res{res}.h5"

    # Load AnnData object
    adata = anndata.read_h5ad(sample_mapping_file)
    res_float = int(res) / 10.0

    # Dimensions
    n_cells = adata.n_obs
    n_genes = adata.n_vars

    print(f"Resolution {res_float:.1f}: {n_cells} cells, {n_genes} genes")

    # Add to summary text for the figure
    summary_lines.append(f"Res {res_float:.1f}: {n_cells} cells, {n_genes} genes")

    # Stats dataframe
    stats = pd.DataFrame({
        "Cell_ID": adata.obs.index.tolist(),
        "Number_of_Cells": adata.obs["Number_of_cells"].tolist(),
        "Sample": adata.obs["scRNA_Sample_Name"].astype(str).tolist(),
    })

    # Average number of cells per pseudocell per sample
    grouped = (
        stats.groupby(["Cell_ID", "Sample"])["Number_of_Cells"]
        .mean()
        .reset_index()
    )

    # Mean number of cells per pseudocell for each sample
    avg_cells_per_pseudocell = (
        grouped.groupby("Sample")["Number_of_Cells"]
        .mean()
        .reset_index()
        .rename(columns={"Number_of_Cells": "Mean_Number_of_Pseudocells"})
    )

    # Total number of cells per donor (Sample)
    total_cells_per_donor = (
        stats.groupby("Sample")["Number_of_Cells"]
        .sum()
        .reset_index()
        .rename(columns={"Number_of_Cells": "Total_Number_of_Cells"})
    )

    # Merge
    merged_stats = total_cells_per_donor.merge(
        avg_cells_per_pseudocell, on="Sample"
    )

    # Scatter for this resolution
    plt.scatter(
        merged_stats["Total_Number_of_Cells"],
        merged_stats["Mean_Number_of_Pseudocells"],
        s=25,
        alpha=0.7,
        color=color,
        label=f"Res {res_float:.1f}",
    )

# Axes labels and title
plt.ylabel("Average Number of Cells per Pseudocell", fontsize=14)
plt.xlabel("Total Number of Cells per Donor", fontsize=14)
plt.title("Metacell Pseudobulk Across Leiden Resolutions", fontsize=14)

plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(title="Leiden resolution", fontsize=9)

# Add dimensions summary as a small text box inside the plot
ax = plt.gca()
ax.text(
    0.05, 0.95,
    "\n".join(summary_lines),
    transform=ax.transAxes,
    va="top",
    ha="left",
    fontsize=8,
    bbox=dict(boxstyle="round", alpha=0.2)
)

plt.tight_layout()

# Save plot
out_pdf = "metacell_pseudobulk_by_resolution.pdf"
out_png = "metacell_pseudobulk_by_resolution.png"

plt.savefig(out_pdf, dpi=300, bbox_inches="tight")
plt.savefig(out_png, dpi=300, bbox_inches="tight")

plt.show()

print(f"Saved plot to: {out_pdf}")
print(f"Saved plot to: {out_png}")
```
